```python
print("******Welcome to online Car Rental Program******")
print("******I am your Virtual Assitant*******")

import datetime

class car_rental:
    def __init__(self,stock):
        self.stock = stock
    def displaystock(self):
        print("Cars Available",self.stock)
        return self.stock
    def rentOnHourlyBasis(self,n):
        if n <= 0:
            print("n must be positive and greater than zero")
        elif n > self.stock:
            print("we don't have that many Cars Available")
        else:
            start = datetime.datetime.now()
            print("you have rented {} cars at {} hours".format(n,start.hour))
            print(start)
            print("you will be charged at Rs100 per hour per car ")
            print("Total price=",n*100)
            print("Total cars remain in stock",self.stock-n)
            print("you has successfully rented car")
            print("We hope that you enjoy our service.")
            print("Thank you!!")
    
    def returnforHourlyBasis(self,n):
        if n<=0:
            print("Enter the positive value or greater than zero")
        elif n>self.stock:
            print("Enter the value less then stock") 
        else:
            end = datetime.datetime.now()
            print("Rental mode = Hourly")
            print("You will be charged 100 for per hour per car.")
            print("Total price=",n*100)
            print(end)
            print("you has successfully returned car")
            print("We hope that you enjoy our service.")
            print("Thank you!!")
            
            
            
    def rentOnDailyBasis(self,n):
        if n <= 0:
            print("n must be positive and greater than zero")
        elif n > self.stock:
            print("we don't have that many Cars Available")
        else:
            start = datetime.datetime.now()
            print("you have rented {} cars at {} hours".format(n,start.hour))
            print(start)
            print("you will be charged at Rs2,000 per car per day ")
            print("Total price=",n*2000)
            print("Total cars remain in stock",self.stock-n)
            print("you has successfully rented car")
            print("We hope that you enjoy our service.")
            print ("Thank you!!")
    def returnforDailyBasis(self,n):
        if n<=0:
            print("Enter the positive value or greater than zero")
        elif n>self.stock:
            print("Enter the value less then stock") 
        else:
            end = datetime.datetime.now()
            print("Rental mode = Daily")
            print("You will be charged 2,000 for each per car per day.")
            print("Total price=",n*2000)
            print(end)
            print("you has successfully returned car")
            print("We hope that you enjoy our service.")
            print("Thank you!!")
        
    def rentOnWeeklyBasis(self,n):
        if n <= 0:
            print("n must be positive and greater than zero")
        elif n > self.stock:
            print("we don't have that many Cars Available")
        else:
            start = datetime.datetime.now()
            print("you have rented {} cars at {} hours".format(n,start.hour))
            print(start)
            print("you will be charged at Rs10,000 per car per week ")
            print("Total price=",n*10000)
            print("Total cars remain in stock",self.stock-n)
            print("you has successfully rented car")
            print("We hope that you enjoy our service.")
            print("Thank you!!")
    def returnforWeeklyBasis(self,n):
        if n<=0:
            print("Enter the positive value or greater than zero")
        elif n>self.stock:
            print("Enter the value less then stock") 
        else:
            end = datetime.datetime.now()
            print("Rental mode = Weekly")
            print("You will be charged 10,000 for each per car per week ")
            print("Total price=",n*10000)
            print(end)
            print("you has successfully returned car")
            print("We hope that you enjoy our service.")
            print("Thank you!!")

while True:   
        obj = car_rental(100)
        userchoice =int(input('''
        What can i help you with today?
        1.Total numbers of available cars
        2.Request a car on hourly basis Rs.100/-
        3.Request a car on daily basis Rs.2,000/-
        4.Request a car on weekly basis Rs.10,000/-
        5.Return car(Hourlybasis) - bill generate
        6.Return car(dailybasis) - bill generate
        7.Return car(weeklybasis) - bill generate
        8.Exit
        '''))
        
        if userchoice==1:
             obj.displaystock()
        elif userchoice==2:
            n=int(input("Enter the quantity:"))
            obj.rentOnHourlyBasis(n)
        elif userchoice==3:
            n=int(input("Enter the quantity:"))  
            obj.rentOnDailyBasis(n)
        elif userchoice==4:
            n=int(input("Enter the quantity:"))
            obj.rentOnWeeklyBasis(n)
        elif userchoice==5:
            n=int(input("Enter the quantity:"))
            obj.returnforHourlyBasis(n)
        elif userchoice==6:
            n=int(input("Enter the quantity:"))
            obj.returnforDailyBasis(n)
        elif userchoice==7:
            n=int(input("Enter the quantity:"))
            obj.returnforWeeklyBasis(n)     
        elif userchoice==8:
            break
        else:
            print("Invalid input. Please enter number between 1-6 ")
```
